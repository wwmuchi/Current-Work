```python
cols_to_keep = [
    'stat_letters',
    'stat_rating',
    'eng_city', 'eng_latitude', 'eng_longitude',
    'stat_format1', 'stat_format2', 'stat_format3',
    'stat_initial_hh_exp'
]
```

```python
column_renaming = {
    'stat_letters' : 'Station Call Sign',
    'stat_rating' : 'Station Rating',
    'eng_city' : 'Antenna City',
    'eng_latitude' : 'Antenna Latitude',
    'eng_longitude' : 'Antenna Longitude',
    'stat_format1' : 'Genre 1',
    'stat_format2' : 'Genre 2',
    'stat_format3' : 'Genre 3',
    'stat_initial_hh_exp' : 'First Year Hip Hop is Played'
}

df = df[cols_to_keep].rename(columns=column_renaming)
df.head()
```
