```python
import pandas as pd
import numpy as np

data = pd.read_csv('Exports/Data/2.StationDataAllYears.csv', index_col=0)
```

```python
# Define the formats and the column for grouping
formats = ['Hip Hop', 'Black', 'Urban Contemporary']

# Initialize a dictionary to store results
group_min_years = {}

# Group the data by the specified column
grouped = data.groupby('letters')

count = 0

for letter, format_data in grouped:

    count += 1
    if count % 1000 == 0:
        print(f'percent complete: {count / len(grouped) * 100:.2f}%')
    
    group_min_years[letter] = {}

    for fmt in formats:
        mask = format_data.apply(
            lambda row: any(fmt == value for value in row),
            axis=1
        )

        filtered_years = format_data.loc[mask, 'year']

        group_min_years[letter][fmt] = filtered_years.min() if not filtered_years.empty else None
    
print(group_min_years)
```

```python
data_to_save = pd.DataFrame(group_min_years).T
assert data_to_save.index.duplicated().sum() == 0
data_to_save.to_csv('Exports/Data/5.InitialExpDates.csv')
data_to_save.head()
```
