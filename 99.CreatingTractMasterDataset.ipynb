***Importing 2010 Outcome Data***

```python
import pandas as pd

filepath = 'Imports/OutcomeData/EducationalAttainment/EducationalAttainment.csv'

outcome_data = pd.read_csv(filepath)
```

```python
col_names = outcome_data.iloc[0,:]

first_col_name = 'Total!!Estimate!!Population 25 to 34 years'
last_col_name = 'Total!!Estimate!!Population 35 to 44 years'

first_col_index = col_names[col_names == first_col_name].index[0]
last_col_index = col_names[col_names == last_col_name].index[0]

col_range = col_names[first_col_index:last_col_index]

cols_to_keep = col_range[~col_range.str.contains("Margin of Error")]
cols_to_keep = cols_to_keep[:-1]

outcome_data = outcome_data.loc[:, ['GEO_ID', 'NAME'] + list(cols_to_keep.index)].copy()
```

```python
assert (outcome_data['Geography'].str.startswith('1400000US')).all()
outcome_data['Geography'] = outcome_data['Geography'].str.replace('1400000US', '')

outcome_data['STATEFP'] = outcome_data['Geography'].str[:2]  # First 2 digits
outcome_data['COUNTYFP'] = outcome_data['Geography'].str[2:5]  # Next 3 digits
outcome_data['TRACTCE'] = outcome_data['Geography'].str[5:]  # Last 6 digits

outcome_data = outcome_data.drop(columns = ['Geography', 'Geographic Area Name'])
outcome_data = outcome_data[['STATEFP', 'COUNTYFP', 'TRACTCE'] + [col for col in outcome_data.columns if col not in ['STATEFP', 'COUNTYFP', 'TRACTCE']]]

assert outcome_data.isna().sum().sum() == 0
```

***Mapping 2000 block groups to 2010 census tracts***

2000 Block Group Parts --> 2000 Block Groups --> 2000 Census Tract

Step 1: crosswalk conversion of bgp --> bg

```python
filepath = 'Imports/Crosswalks/CensusTract/nhgis_bgp2000_tr2010/nhgis_bgp2000_tr2010.csv'
crosswalk = pd.read_csv(filepath)
```

```python
block_group_part_gisjoin_2000 = crosswalk['bgp2000gj']
assert (block_group_part_gisjoin_2000.apply(len).unique() == 26)

def convert_to_15char_gisjoin(gisjoin):
    
    state_code = gisjoin[1:4]       # State NHGIS code (3 digits)
    county_code = gisjoin[4:8]      # County NHGIS code (4 digits)
    tract_code = gisjoin[18:24]     # Census tract code (6 digits)
    block_group = gisjoin[25]       # Block group code (1 digit)
    
    return f"G{state_code}{county_code}{tract_code}{block_group}"

block_group_gisjoin_2000 = block_group_part_gisjoin_2000.apply(convert_to_15char_gisjoin)

crosswalk['bg2000gj'] = block_group_gisjoin_2000
```

```python
print('every GISJOIN in df147 and df152 match an entry under "bg2000gj" in the crosswalk:')
assert df147['GISJOIN'].isin(crosswalk['bg2000gj']).all()
assert df152['GISJOIN'].isin(crosswalk['bg2000gj']).all()
```

```python
# renaming and reordering
crosswalk = crosswalk[['bg2000gj', 'bgp2000gj', 'tr2010gj'] + [col for col in crosswalk.columns if 'wt' in col]].copy()
crosswalk.rename(columns={col: f'bgp_{col}' for col in crosswalk.columns if 'wt' in col}, inplace=True)
```

```python
for col in crosswalk.columns:
    if 'wt' in col:
        assert (crosswalk.groupby('bgp2000gj')[col].sum().apply(lambda x: round(x,2)) == 1).all()
        assert not (crosswalk.groupby('bg2000gj')[col].sum().apply(lambda x: round(x,2)) == 1).all()

print('bgp weights sum to 1 across each block group part')
print('bgp weights do not sum to 1 each block group')
```

```python
# converting weights from block group part to block group level
weight_cols = [col for col in crosswalk.columns if 'wt' in col]
print(weight_cols)

for col in weight_cols:
    new_col_name = col.replace('bgp', 'bg')
    per_bg = crosswalk.groupby('bg2000gj')[col].transform('sum')
    per_bg_tract = crosswalk.groupby(['bg2000gj', 'tr2010gj'])[col].transform('sum')
    crosswalk[new_col_name] = per_bg_tract / per_bg

crosswalk.drop_duplicates(subset=['bg2000gj', 'tr2010gj'], inplace=True)
crosswalk.drop(columns=weight_cols + ['bgp2000gj'], inplace=True)
weight_cols = [col for col in crosswalk.columns if 'wt' in col]
print(weight_cols)

for col in weight_cols:
    assert (crosswalk.groupby('bg2000gj')[col].sum().apply(lambda x: round(x,2)) == 1).all(), f'column {col} does not sum to 1'
    
print('bg weights sum to 1 across each block group')
```

Step 2. df147 conversion from bg20000 to tr2010

```python
assert not df147['GISJOIN'].duplicated().any()
print(f'each row refers to a unique block group in df147')
assert df147['GISJOIN'].isin(crosswalk['bg2000gj']).all()
print(f'every row has a match in the crosswalk')

data_cols = [col for col in df147.columns if 'FY' in col]

assert df147[data_cols].notnull().all().all()
assert (df147[data_cols].dtypes == 'int64').all()

df147 = df147[['GISJOIN'] + data_cols].copy()
df147.head()
```

All vars are categorized as "Universe: Persons"

```python
df147 = df147.merge(crosswalk[['bg2000gj', 'tr2010gj', 'bg_wt_pop']], left_on='GISJOIN', right_on='bg2000gj', how='left', validate='1:m', indicator=True)
assert df147['_merge'].eq('both').all()
df147.drop(columns=['_merge', 'bg2000gj'], inplace=True)

df147[data_cols] = df147[data_cols].multiply(df147['bg_wt_pop'], axis='rows')
df147.drop(columns='bg_wt_pop', inplace=True)

df147 = df147.groupby('tr2010gj')[data_cols].sum()
df147 = df147.astype(int)
df147.reset_index(inplace=True)
df147.head()
```

Step 3. df152 conversion

```python
assert not df152['GISJOIN'].duplicated().any()
print(f'each row refers to a unique block group in df152')
assert df152['GISJOIN'].isin(crosswalk['bg2000gj']).all()
print(f'every row has a match in the crosswalk')

data_codes = ['HB', 'HDX', 'HEZ', 'HG4', 'G8V']
data_cols = [col for col in df152.columns if any(code in col for code in data_codes)]
df152 = df152[['GISJOIN'] + data_cols].copy()
df152.head()
```

Var categorization:
- ADULT POPULATION:
  - Population 15 Years and Over by Sex by Marital Status
  - Population 15 Years and Over Now Married by Sex by Now Married Marital Status
  - Population 15 Years and Over Now Married with Spouse Absent by Sex by Spouse Absent Type
  - Population 16 Years and Over by Sex by Employment Status
  - Per Capita Income in 1999
- FAMILIES
  - Population 3 Years and Over by Sex by School Enrollment by School Type
- OWNER-OCCUPIED HOUSING UNITS
  - Specified Owner-Occupied Housing Units

```python
weight_type_dict = {'bg_wt_adult' : ['HB', 'HEZ', 'HG4'], 'bg_wt_fam': ['HDX'], 'bg_wt_ownhu': ['G8V']}

df152 = df152.merge(crosswalk[['bg2000gj', 'tr2010gj'] + list(weight_type_dict.keys())], left_on='GISJOIN', right_on='bg2000gj', how='left', validate='1:m', indicator=True)
assert df152['_merge'].eq('both').all()
df152.drop(columns=['_merge', 'bg2000gj'], inplace=True)

for weight_type, data_codes in weight_type_dict.items():
    data_cols = [col for col in df152.columns if any(code in col for code in data_codes)]
    df152[data_cols] = df152[data_cols].multiply(df152[weight_type], axis='rows')
    df152.drop(columns=weight_type, inplace=True)
```

```python
data_columns = [col for col in df152.columns if col not in ['GISJOIN', 'tr2010gj']]
print('cols to sum:' , data_columns)
df152 = df152.groupby('tr2010gj')[data_columns].sum()
df152 = df152.astype(int)
df152.reset_index(inplace=True)
df152.head()
```

Step 5b. Normalizing counts

```python
# Normalize counts to percentages
column_groups = {
    'Population by Race': [f'FYE{str(i).zfill(3)}' for i in range(1, 8)],
    # 'Population by Age': [col for col in data.columns if 'age' in col],
    'Population by Age': [f'FYM{str(i).zfill(3)}' for i in range(1, 24)],
    'Population by Sex by Marital Status': [f'HB7{str(i).zfill(3)}' for i in range(1, 9)],
    'Population Now Married by Sex by Now Married Marital Status': [f'HB8{str(i).zfill(3)}' for i in range(1, 5)],
    'Population Now Married with Spouse Absent by Sex by Spouse Absent Type': [f'HB9{str(i).zfill(3)}' for i in range(1, 5)],
    'Population by Sex by Employment Status': [f'HEZ{str(i).zfill(3)}' for i in range(1, 5)]
}

pct_cols = [col for sublist in column_groups.values() for col in sublist]
print(pct_cols)

raw_cols = [col for col in data.columns if col not in pct_cols and col != 'tr2010gj']
print(raw_cols)

assert set(data.columns).difference(set(pct_cols).union(set(raw_cols))) == {'tr2010gj'}
print('all data columns are accounted for')
```

```python
for group, cols in column_groups.items():
    total = data[cols].sum(axis=1)
    total = total.replace(0, 1) # so we don't divide by zero
    for col in cols:
        data[col] = round((data[col] / total) * 100, 3)

assert data.isnull().sum().sum() == 0
assert data[pct_cols].apply(lambda x: x.between(0, 100).all()).all()

data[pct_cols].head()
```

```python
for k,v in column_groups.items():
    print(k)
    print(v)
```

```python
# Determining number of clusters using elbow method
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.preprocessing import StandardScaler

type_to_view = 'Population Now Married by Sex by Now Married Marital Status'

group_data = data[column_groups[type_to_view]].copy()

scaler = StandardScaler()
group_data_scaled = scaler.fit_transform(group_data)

wcss = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(group_data_scaled)
    wcss.append(kmeans.inertia_)

# Plot the elbow graph
plt.plot(range(1, 11), wcss, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('Within-Cluster Sum of Squares (WCSS)')
plt.title(f'Elbow Method: {type_to_view}')
plt.show()
```

```python
optimal_clusters = 3

kmeans = KMeans(n_clusters=optimal_clusters, random_state=42)
kmeans.fit(group_data_scaled)

group_data['Cluster'] = kmeans.labels_

cluster_summary = group_data.groupby('Cluster').mean()

cluster_summary
```

```python
# Plot each row as a separate bar chart
for index, row in cluster_summary.iterrows():
    plt.figure(figsize=(10, 5))  # Create a new figure for each cluster
    
    # Create bar chart for the current row
    plt.bar(cluster_summary.columns, row.values, label=f'Cluster {index}', alpha=0.7)
    
    plt.xlabel('Features (Columns)')
    plt.ylabel('Values')
    plt.title(f'Cluster {index} Summary Bar Chart')
    plt.xticks(rotation=90)
    plt.legend()
    plt.tight_layout()
    plt.show()
```

```python
numb_clusters = {
    'Population by Race': 4,
    'Population by Age': 2,
    'Population by Sex by Marital Status': 3,
    'Population Now Married by Sex by Now Married Marital Status': 3,
    'Population Now Married with Spouse Absent by Sex by Spouse Absent Type': 3,
    'Population by Sex by Employment Status': 3
}

assert set(column_groups.keys()) == set(numb_clusters.keys())

cluster_results = pd.DataFrame(index=data.index)

for name, group_columns in column_groups.items():
    print(name)
    print(group_columns)

    group_data = data[group_columns].copy()

    # Standardize the data
    scaler = StandardScaler()
    group_data_scaled = scaler.fit_transform(group_data)

    # Apply K-Means Clustering
    kmeans = KMeans(n_clusters=numb_clusters[name], random_state=42)
    kmeans.fit(group_data_scaled)

    cluster_results[f'Cluster: {name}'] = kmeans.predict(group_data_scaled)

data2 = pd.concat([data, cluster_results], axis=1)
cluster_cols = [col for col in data2.columns if 'Cluster' in col]
data2[cluster_cols].head()
```

```python
# Compute descriptive statistics for each cluster
name_to_investigate = 'Population by Age'

grouped_data = data2.groupby(f'Cluster: {name_to_investigate}')[column_groups[name_to_investigate]]

cluster_mean = grouped_data.mean().apply(lambda x: round(x, 2))
cluster_std = grouped_data.std().apply(lambda x: round(x, 2))
cluster_count = grouped_data.size()

cluster_mean
```
